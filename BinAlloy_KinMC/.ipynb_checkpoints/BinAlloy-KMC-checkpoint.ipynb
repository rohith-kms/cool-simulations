# Kinetic MC Simualtion of vacancy in 2D binary alloy

## Some binary alloy basics:

A binary alloy is an alloy with two components. Let's call the components A and B. At a particular temperature and pressure, let's denote the molar Gibbs free energies of pure A and Pure B as $G_A$ and $G_B$ respectively.

When we make an alloy out of the two components with mole fractions $X_A$ and $X_B$ (note that $X_A = 1-X_B$, since it's a binary alloy) and assume complete random mixing, the molar Gibbs free energy will correspond to:

$$G = X_A G_A + X_B G_B + \Delta G_{mix}$$

Where $\Delta G_{mix} = \Delta H_{mix} -T \Delta S_{mix}$

A constant temperature and pressure system will try to decrease its molar Gibbs free energy, and the equilibrium configuration corresponds to the one with least Gibbs free energy. At higher temperatures the free energy of the disordered (completely random mixing) alloy decreases, and therefore at higher temperatures we expect binary alloys to be more disordered.

At lower temperatures however, $\Delta H_{mix}$ determines the equilibrium configuration. If $\Delta H_{mix} <0$ mixing is favored and the system tries to maximize the number of A-B bonds, leading to an ordered system. If $\Delta H_{mix} > 0$, system tries to minimize the number of A-B bonds, leading to the formation of A-rich and B-rich clusters.

If we use a nearest neighbor bond model, with A-A,B-B and A-B bond energies given by $\phi_{AA}$, $\phi_{BB}$, and $\phi_{AB}$ respectively, the enthalpy of mixing can be given by:

$$\Delta H_{mix} = \Omega X_A X_B$$

where $$\Omega = \phi_{AB} - \frac{1}{2}(\phi_{AA} + \phi_{BB}) $$

## Vacancies and Diffusion

When an atom is missing in a lattice site of a crystal, it's called a vacancy. Vacancies are point defects. Vacancies aid the transport of atoms within the crystal (diffusion) by providing a low-energy site next to an atom where it can jump, given enough vibrational energy. In the absence of other defects such as dislocations and grain boundaries, vacancies mediate diffusion in crystals.

Since vacancies are equilibium defects, there is always a finite concentration of vacancies in a crystal above 0K.

Our crystal will be a square lattice of size $n \times n$ with each site occupied by an A or B atom. We will impose periodic boundary conditions. Let's define some essential functions that will help our simulation.

## Essential Functions

```python
# Let's define some necessary functions

# Our crystal of size 'n' will be represented by a n x n matrix of ones and zeros. (called lat)
# 0,1 correspond to A and B atoms respectively
# -1 corresponds to a vacancy in the crystal 

def periodic_bd(lat):
    
    n = lat.shape[0]
    larger_lat = np.ones([n+2,n+2])*0.5
    larger_lat[1:-1,1:-1] = lat
    larger_lat[1:-1,0] = lat[:,-1]
    larger_lat[1:-1,-1] = lat[:,0]
    larger_lat[0,1:-1] = lat[-1,:]
    larger_lat[-1,1:-1] = lat[0,:]
    
    return larger_lat

def calc_energy_total(lat):
    
    larger_lat = periodic_bd(lat)
    energy=0
    
    for i in range(n+1):        
        for j in range(n+1):
            
            energy += (larger_lat[i,j]==larger_lat[i+1,j]==0)*pAA + \
            (larger_lat[i,j]==larger_lat[i+1,j]==1)*pBB + \
            ((larger_lat[i,j]!=larger_lat[i+1,j])&(larger_lat[i+1,j]!=0.5))*pAB + \
            (larger_lat[i,j]==larger_lat[i,j+1]==0)*pAA + \
            (larger_lat[i,j]==larger_lat[i,j+1]==1)*pBB + \
            ((larger_lat[i,j]!=larger_lat[i,j+1])&(larger_lat[i,j+1]!=0.5))*pAB
            
    return energy

def calc_energy_one(pos,larger_lat):
    
    i1 = pos[0]
    j1 = pos[1]
    
    # wrapping to avoid array out of bounds
    if (j1<1):    
        j1=n
        
    if (j1>n):
        j1=1
    
    if (i1<1):    
        i1=n
        
    if (i1>n):
        i1=1
    
    energy = (larger_lat[i1,j1]==larger_lat[i1+1,j1]==0)*pAA + \
    (larger_lat[i1,j1]==larger_lat[i1+1,j1]==1)*pBB + \
    ((larger_lat[i1,j1]!=larger_lat[i1+1,j1])&(larger_lat[i1+1,j1]!=0.5))*pAB + \
    (larger_lat[i1,j1]==larger_lat[i1,j1+1]==0)*pAA + \
    (larger_lat[i1,j1]==larger_lat[i1,j1+1]==1)*pBB + \
    ((larger_lat[i1,j1]!=larger_lat[i1,j1+1])&(larger_lat[i1,j1+1]!=0.5))*pAB + \
    (larger_lat[i1,j1]==larger_lat[i1,j1-1]==0)*pAA + \
    (larger_lat[i1,j1]==larger_lat[i1,j1-1]==1)*pBB + \
    ((larger_lat[i1,j1]!=larger_lat[i1,j1-1])&(larger_lat[i1,j1-1]!=0.5))*pAB + \
    (larger_lat[i1,j1]==larger_lat[i1-1,j1]==0)*pAA + \
    (larger_lat[i1,j1]==larger_lat[i1-1,j1]==1)*pBB + \
    ((larger_lat[i1,j1]!=larger_lat[i1-1,j1])&(larger_lat[i1-1,j1]!=0.5))*pAB
    
    return energy

def swap(pos1,pos2,larger_lat):
    
    #pos1 is vacancy position
    new = np.copy(larger_lat)
    temp = new[pos2]
    new[pos2] = new[pos1]
    new[pos1] = temp
    
    delta_e = (-calc_energy_one(pos1,larger_lat) \
    - calc_energy_one(pos2,larger_lat) \
    + calc_energy_one(pos1,new) \
    + calc_energy_one(pos2,new))
    
    return new,delta_e
```

# Kinetic Monte Carlo algorithm:

Kinetic Monte Carlo algorithm (KMC) also known as the residence time algorithm or the BKL algorithm, is a Monte Carlo method to simulate time evolution. The steps in the algorithm are as follows:

1. Start with an initial state $k$ at time $t = 0$
2. List out the rates of transition from $k$ to a state $i$ as $r_{ki}$ for all $N_k$ possible states of transition in the system
3. Calculate the cumulative function $R_{ki} = \sum_{j=1}^{i} r_{kj}$ for $i=1,2...N_k$. Total rate $Q_k = R_{k,N_k}$
4. Pick a random number $u \in (0,1]$ and choose the succesive event $i$ such that $R_{k,i-1} < uQ_k \leq R_{ki}$
5. Carry out the transition $k \rightarrow i$
6. Update time $t$ to $t+\Delta t$ such that $\Delta t = Q_k^{-1} ln{\frac{1}{u'}}$ where $u'$ is a random number such that $u' \in (0,1]$
7. Return to step 3

In this manner, the time evolution of the sustem can be obtained if we can calculate the rates for all possible transition states given a current state.

# Simulation step:

In our simulation, given a state there are 4 possible transition states - one for each neighbor of the vacancy. The rates are calculated as 

$$r_{ki} = e^{\frac{-\Delta H_{ki}}{\kappa T}}$$

where $\kappa$ is the boltzmann parameter, and $T$ is the temperature.

$\Delta H_{ki}$ is calculated for $i=1,2,3,4$ by swapping the vacancy with one of its ($i^{th}$) nearest neighbors and computing the difference in energies unsing the **swap()** function. Then, the KMC algorithm given above is applied for a given number of iterations.

```python
def iteration(larger_lat,vacancy_pos,kT):
    
    # Finding different paths and their relative kinetics
    
    up = vacancy_pos[0]-1,vacancy_pos[1]
    down = vacancy_pos[0]+1,vacancy_pos[1]
    right = vacancy_pos[0],vacancy_pos[1]+1
    left = vacancy_pos[0],vacancy_pos[1]-1
    
    lat_up,e_up = swap(vacancy_pos,up,larger_lat)
    lat_down,e_down = swap(vacancy_pos,down,larger_lat)
    lat_right,e_right = swap(vacancy_pos,right,larger_lat)
    lat_left,e_left = swap(vacancy_pos,left,larger_lat)
    
    positions = [up,down,right,left]
    energies = np.array([e_up,e_down,e_right,e_left])
    rates = np.exp(-energies/kT)
    lats = [lat_up,lat_down,lat_right,lat_left]
    
    # Choosing which path to take
    
    cum_rates = np.cumsum(rates)
    rand = np.random.rand()*cum_rates[-1]
    index = np.where(np.array(cum_rates>rand) ==True)[0][0]
    
    # Compute timestep (delta_t)
    delta_t = np.log(1/np.random.rand())/cum_rates[-1]
    
    # Apply periodic boundary conditions.
    
    if (positions[index][0]<1): 
        
        x = n
        y = positions[index][1]
        positions[index] = x,y
        lats[index][positions[index]] = -1
        
    if (positions[index][1]<1): 
        
        x = positions[index][0]
        y = n
        positions[index]= x,y
        lats[index][positions[index]] = -1
        
    if (positions[index][0]>n): 
        
        x = 1
        y = positions[index][1]
        positions[index] = x,y
        lats[index][positions[index]] = -1
    
    if (positions[index][1]>n): 
        
        x = positions[index][0]
        y = 1
        positions[index]= x,y
        lats[index][positions[index]] = -1
        
    lats[index] = periodic_bd(lats[index][1:-1,1:-1])
    
    # return the vacancy position, new lattice and change in energy
    
    return lats[index], positions[index], energies[index], delta_t
```

# Ordering system ( $\Omega < 0$ ):

We choose

* $\phi_{AA} = -1 eV$
* $\phi_{BB} = -1 eV$
* $\phi_{AB} = -1.5 eV$

Giving us $\Omega = -0.5 eV$. We choose a $12 \times 12$ lattice, with a random, 50% distribution of A and B and run the simulation for 10,000 KMC steps. We choose kT = 0.1 which corresponds to a temperature of about 1160 K. 

```python
# Define the bond energies

pAA = -1
pBB = -1
pAB = -1.5

# lattice initialisation
n=12

lat = (np.random.rand(n,n)<0.5)*1.0
init_lat=lat

vacancy_pos = 2,2
vacancy_pos = vacancy_pos[0]+1,vacancy_pos[1]

lat[vacancy_pos] = -1

larger_lat = np.ones([n+2,n+2])*0.5
larger_lat = periodic_bd(lat)
vacancy_pos = vacancy_pos[0]+1,vacancy_pos[1]+1



total_time = 10000
record = np.zeros([total_time,n,n])
energies = np.zeros(total_time)
time = np.zeros(total_time)
current_energy=0
current_time = 0

kT = 0.1     


for i in range(total_time):
    
    global larger_lat
    record[i,:,:] = larger_lat[1:-1,1:-1]
    energies[i] = current_energy
    time[i] = current_time
    larger_lat,vacancy_pos,delta_e,delta_t = iteration(larger_lat,vacancy_pos,kT)
    current_energy+= delta_e
    current_time+= delta_t
    
    
# record[i,:,:] has the state for the i th KMC step
# energies[i] has the energy and time[i] has the timestep for the i th KMC timestep
```
